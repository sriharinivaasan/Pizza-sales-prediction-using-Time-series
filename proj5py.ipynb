```python
weekly_orders_all
```

```python
# Prepare a list to hold ADF results
adf_results = []

# Loop through each unique pizza
for pizza in weekly_orders_all['pizza_name_id'].unique():
    # Get the time series for that pizza
    ts = weekly_orders_all[weekly_orders_all['pizza_name_id'] == pizza]['order_count']
    
    # Apply the ADF test
    try:
        result = adfuller(ts)
        adf_stat = result[0]
        p_value = result[1]
        
        
        

    except:
        adf_stat = None
        p_value = None

    # Store results
    adf_results.append({
        'pizza_name_id': pizza,
        'adf_statistic': adf_stat,
        'p_value': p_value
    })

# Convert to DataFrame
adf_results_df = pd.DataFrame(adf_results)
adf_results_df['p_value_fixed'] = adf_results_df['p_value'].apply(lambda x: f"{x:.15f}")
# Show the first few results
print(adf_results_df)
```

```python
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_percentage_error
import matplotlib.pyplot as plt
import pandas as pd

# Ensure datetime
combined_df['order_date'] = pd.to_datetime(combined_df['order_date'])

# Initialize list to collect MAPE values
mape_list = []

# Get all unique pizza IDs
pizza_ids = combined_df['pizza_name_id'].unique()

for pizza_id in pizza_ids:
    df = combined_df[combined_df['pizza_name_id'] == pizza_id].copy()
    
    # Skip if too few records
    if len(df) < 10:
        continue

    # Prepare time series
    df.set_index('order_date', inplace=True)
    df = df.asfreq('W')
    df = df.sort_index()
    
    # Drop NaNs (from asfreq)
    df = df.dropna()
    
    # Skip if too few rows after drop
    if len(df) < 10:
        continue

    # Train-test split (last 4 weeks as test)
    train = df.iloc[:-4]
    test = df.iloc[-4:]

    try:
        # Auto ARIMA to determine best (p,d,q)
        stepwise_model = auto_arima(train['order_count'], seasonal=False, trace=False,
                                    error_action='ignore', suppress_warnings=True,
                                    max_p=5, max_q=5, max_d=2, stepwise=True)
        
        model = ARIMA(train['order_count'], order=stepwise_model.order)
        fitted_model = model.fit()
        forecast = fitted_model.forecast(steps=len(test))

        # Calculate MAPE
        mape = mean_absolute_percentage_error(test['order_count'], forecast)
        mape_list.append((pizza_id, mape))
        print(f"Pizza: {pizza_id}, MAPE: {mape:.2%}")

    except Exception as e:
        print(f"Skipped {pizza_id} due to error: {e}")

# Average MAPE
if mape_list:
    avg_mape = sum(m[1] for m in mape_list) / len(mape_list)
    print(f"\nAverage MAPE across {len(mape_list)} pizzas: {avg_mape:.2%}")
else:
    print("No valid pizzas for ARIMA evaluation.")
```

* SARIMA

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_absolute_percentage_error
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Ensure datetime
combined_df['order_date'] = pd.to_datetime(combined_df['order_date'])

mape_list = []
pizza_ids = combined_df['pizza_name_id'].unique()

for pizza_id in pizza_ids:
    df = combined_df[combined_df['pizza_name_id'] == pizza_id].copy()
    
    if len(df) < 20:
        continue

    df.set_index('order_date', inplace=True)
    df = df.sort_index().asfreq('W')

    df = df.dropna()

    if len(df) < 20:
        continue

    train = df.iloc[:-4]
    test = df.iloc[-4:]

    try:
        model = SARIMAX(train['order_count'],
                        order=(1, 1, 1),
                        seasonal_order=(1, 1, 1, 52),
                        enforce_stationarity=False,
                        enforce_invertibility=False)
        
        results = model.fit(disp=False)
        forecast = results.forecast(steps=4)
        mape = mean_absolute_percentage_error(test['order_count'], forecast)
        mape_list.append((pizza_id, mape))

        print(f"Pizza: {pizza_id}, MAPE: {mape:.2%}")
        
    except Exception as e:
        print(f"Skipped {pizza_id} due to error: {e}")

# Average MAPE
if mape_list:
    avg_mape = sum(m[1] for m in mape_list) / len(mape_list)
    print(f"\nAverage MAPE across {len(mape_list)} pizzas: {avg_mape:.2%}")
else:
    print("No valid pizzas for SARIMA evaluation.")
```

* Prophet

```python
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Ensure datetime
combined_df['order_date'] = pd.to_datetime(combined_df['order_date'])

mape_list = []
pizza_ids = combined_df['pizza_name_id'].unique()

for pizza_id in pizza_ids:
    df = combined_df[combined_df['pizza_name_id'] == pizza_id].copy()
    
    if len(df) < 20:
        continue

    df = df[['order_date', 'order_count']].rename(columns={'order_date': 'ds', 'order_count': 'y'})
    df = df.sort_values('ds')

    df = df.set_index('ds').asfreq('W').dropna().reset_index()

    if len(df) < 20:
        continue

    train = df.iloc[:-4]
    test = df.iloc[-4:]

    try:
        model = Prophet(weekly_seasonality=True, yearly_seasonality=True, daily_seasonality=False)
        model.fit(train)

        future = model.make_future_dataframe(periods=4, freq='W')
        forecast = model.predict(future)

        predicted = forecast[['ds', 'yhat']].set_index('ds').loc[test['ds']].yhat.values
        actual = test['y'].values

        mape = mean_absolute_percentage_error(actual, predicted)
        mape_list.append((pizza_id, mape))

        print(f"Pizza: {pizza_id}, MAPE: {mape:.2%}")

    except Exception as e:
        print(f"Skipped {pizza_id} due to error: {e}")

# Average MAPE
if mape_list:
    avg_mape = sum(m[1] for m in mape_list) / len(mape_list)
    print(f"\nAverage MAPE across {len(mape_list)} pizzas: {avg_mape:.2%}")
else:
    print("No valid pizzas for Prophet evaluation.")
```

* LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Ensure datetime
combined_df['order_date'] = pd.to_datetime(combined_df['order_date'])

mape_list = []
pizza_ids = combined_df['pizza_name_id'].unique()

for pizza_id in pizza_ids:
    df = combined_df[combined_df['pizza_name_id'] == pizza_id].copy()
    
    if len(df) < 20:
        continue

    df = df[['order_date', 'order_count']].sort_values('order_date')
    df.set_index('order_date', inplace=True)
    df = df.asfreq('W').dropna()

    # Create lag features
    df['lag_1'] = df['order_count'].shift(1)
    df['lag_2'] = df['order_count'].shift(2)
    df['lag_3'] = df['order_count'].shift(3)
    df.dropna(inplace=True)

    if len(df) < 10:
        continue

    # Split into train and test (last 4 weeks as test)
    train = df.iloc[:-4]
    test = df.iloc[-4:]

    X_train = train[['lag_1', 'lag_2', 'lag_3']]
    y_train = train['order_count']
    X_test = test[['lag_1', 'lag_2', 'lag_3']]
    y_test = test['order_count']

    try:
        model = LinearRegression()
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)

        mape = mean_absolute_percentage_error(y_test, predictions)
        mape_list.append((pizza_id, mape))
        print(f"Pizza: {pizza_id}, MAPE: {mape:.2%}")

    except Exception as e:
        print(f"Skipped {pizza_id} due to error: {e}")

# Average MAPE
if mape_list:
    avg_mape = sum(m[1] for m in mape_list) / len(mape_list)
    print(f"\nAverage MAPE across {len(mape_list)} pizzas: {avg_mape:.2%}")
else:
    print("No valid pizzas for Regression evaluation.")
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

# Ensure datetime format
combined_df['order_date'] = pd.to_datetime(combined_df['order_date'])

# Prepare result list
results = []

# Get unique pizza IDs
pizza_ids = combined_df['pizza_name_id'].unique()

for pizza_id in pizza_ids:
    df = combined_df[combined_df['pizza_name_id'] == pizza_id].copy()
    
    # Filter and prepare data
    df = df[['order_date', 'order_count']].sort_values('order_date')
    df.set_index('order_date', inplace=True)
    df = df.asfreq('W').dropna()

    # Create lag features
    df['lag_1'] = df['order_count'].shift(1)
    df['lag_2'] = df['order_count'].shift(2)
    df['lag_3'] = df['order_count'].shift(3)
    df.dropna(inplace=True)

    if len(df) < 10:
        continue

    X = df[['lag_1', 'lag_2', 'lag_3']]
    y = df['order_count']

    try:
        # Fit model
        model = LinearRegression()
        model.fit(X, y)

        # Prepare input for next week prediction
        last_row = df.iloc[-1]
        next_input = np.array([[last_row['order_count'], last_row['lag_1'], last_row['lag_2']]])

        # Predict
        prediction = model.predict(next_input)[0]
        prediction = max(0, round(prediction))  # Ensure non-negative and round

        # Store result
        results.append({'pizza_name_id': pizza_id, 'predicted_order_count': prediction})

    except Exception as e:
        print(f"Skipped {pizza_id} due to error: {e}")

# Create result DataFrame
prediction_df = pd.DataFrame(results)
print(prediction_df)
```
